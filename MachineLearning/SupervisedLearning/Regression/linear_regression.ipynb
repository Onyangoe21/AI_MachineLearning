```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('../../Utilities/styling.mplstyle') # This is just to do the styling
```

Assuming you wanted a linear regression to predict the ounces of water consumed by a given number of people in a day and you know that 20 people consume 1280 ounces in a day and 100 people consume 6400 ounces in a day, your job will be quite simple:
* Indentify your dataset:
* Since you want your model to predict based on known y's for specific x's you can split your data so that you have x's in one list and their corresponding y's in another list to make it easier to process the value pairs as follows:

```python
x_train = np.array([20, 100, 200])
y_train = np.array([1280, 6400, 12800])
print(f"x_train = {x_train}")
print(f"y_train = {y_train}")
```

In this case: x<sup>0</sup>, y<sup>0</sup> = 20, 1280 and x<sup>1</sup>, y<sup>1</sup> = 100, 6400

We know that we need some way to loop through this training examples while doing something to help us come up with the line(in this case our function) and it is important to know how to access values within the data set which has now been reduced to a list of x's and a list of y's.

To access a specific x<sup>i</sup>, we can simply do: 

```python
i = 0
value_accessed_at_i = x_train[i]
print("Value x and index", i, "is", value_accessed_at_i)

# and the same goes for the y values
value_accessed_at_i = y_train[i]
print("Value y and index", i, "is",value_accessed_at_i)
```

### Plotting this Data
* We can now plot the data after knowing how to access it:

```python
# Plot the data points
plt.scatter(x_train, y_train, marker='x', c='r')
# Set the title
plt.title("Water Drank @day in Ounces")
# Set the y-axis label
plt.ylabel('Number of people')
# Set the x-axis label
plt.xlabel('Ounces of water consumed')
plt.show()
```

To have our model make predictions on data points it may not have(the whole point of machine learning), we need to come up with the function that can help us make the prediction for values of x<sup>*i*</sup>. In linear regression this function is linear and thus, our equation is in the format: f<sub>w,b</sub>(x<sup>*i*</sup>) = w(x<sup>*i*</sup>) + b

In this linear function, we have two variables(parameters): w and b. The other value is plugged in from the data points we have. Our goal is to make w(x<sup>*i*</sup>) + b close to our real y values as much as we can. This would ensure that given a new x value, our model can make a prediction value as close to our real y value as possible. You might already be thinking that you are minimizing errors that your function returns and that is exactly what you are doing. 

If you are wondering what is going on with the points we plotted earlier and how that relates to all of this, what we plotted earlier was just a visualization of 2 points to show how we can do plotting so that we can do live experimentations like this one with different values of w and b:

```python

# let us wrap our plotting lines into one function that does the plotting
def plot_values(x_train, y_train, title, y_label, x_label):
    # Plot the data points
    plt.scatter(x_train, y_train, marker='o', c='r')
    # Set the title
    plt.title(title)
    # Set the y-axis label
    plt.ylabel(y_label)
    # Set the x-axis label
    plt.xlabel(x_label)
    plt.show()

title = "Water Drank @day in Ounces"
y_label = 'Number of people'
x_label = 'Ounces of water consumed'

plot_values(x_train, y_train, title, x_label, y_label)
```
